# Agregación  de datos por categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = ['Male', 'Female'] # Genero
income = ['Poor', 'Middle Class', 'Rich'] # NIvel de ingresos
```

```python
## Para generar aleatoriamente los datos
n = 500 ## 500 datos pra cada uno

gender_data = []
income_data = []

for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[:10]
```

```python
income_data[:10]
```

```python
## Z -> N(0,1)
## N(m,s) -> m + s * Z
height = 160 + 30 * np.random.randn(n) ## suelen darse con una distribución normal sumandole a la media(160) una dispersión (30) multiplicada por un número de la normal del tamaño de nuestro dataset (n)
weight = 65 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 * np.random.randn(n)
```

```python
data = pd. DataFrame(
{
    'Gender': gender_data,
    'Economic Status': income_data,
    'Height': height,
    'Weight': weight,
    'Age': age,
    'Income': income
})
```

```python
data.head()
```

### Agrupación de datos por categorías

```python
## Agrupo por categorias de hombre y mujer. Damndome como resultado el índice donde se encuentran cada uno
grouped_gender = data.groupby('Gender') ## Se le asigna a una variable ya que es un objeto de pandas, esto para poder manipularlo
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
## si solo quiero una categoría
grouped_gender.get_group('Female') # Me muestra solo la categoria de mujeres
```

```python
double_group = data.groupby(['Gender', 'Economic Status']) # Agrupando por dos categorias mediante un array en donde pongo las categorias que quiero
```

```python
len(double_group)
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Operaciones sobre datos agrupados

```python
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
grouped_income = double_group['Income'] ## Puedo seleccionar una sola de las columnas del grupo en etse caso 'Income'
```

```python
grouped_income.describe() # Puedo realizar operacioens con esta columna
```

```python
double_group.aggregate(
{
    'Income': np.sum,
    'Age': np.mean,
    'Height': np.std
})
```

```python
double_group.aggregate(
{
    'Age': np.mean,
    'Height': lambda h:(np.mean(h))/np.std(h) # para pasar una función como parametro
})
```

```python
double_group.aggregate([np.sum, np.mean,np.std]) # aplico directamente suma, media y desviación estandar
```

```python
double_group.aggregate([lambda x: np.mean(x) / np.std(x)])
```

## Filtrado de datos

```python
double_group['Age'].filter(lambda x: x.sum()>2400)
```

## Transformación de variables

```python
zscore = lambda x: (x - x.mean())/x.std()
```

```python
import matplotlib.pyplot as plt
z_group =double_group.transform(zscore)
```

```python
plt.hist(z_group['Age'])
```

```python
fill_na_mean = lambda x : x.fillna(x.mean()) # Relleno los NaN con el promedio
```

```python
double_group.transform(fill_na_mean) ## se reemplazaria culaquier valor con NaN con el promedio
```
