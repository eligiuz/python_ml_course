# Carga de datos a través de la función open

```python
data3 = open (mainpath + "/" + "customer-churn-model/Customer Churn Model.txt", "r") # indicar si es modo lectura (r) o lectura escritura (rw)
```

```python
cols = data3.readline().strip().split(',') # con strip elimino cualquier espacio que haya quedado y con split le digo que el separador sera la coma (,)
n_cols = len(cols) # Cuento cuantas columnas son
```

```python
# Con este código puedo encontrar y separar cada columna
counter = 0
main_dict = {} # Se crea un diccionario vacio con las llaves
for col in cols:
    main_dict[col] = [] # Se define el diccionario usando para la clave el nombre de la columna y los valores se le iran agregandoen la siguiente celda
# si llamo a main_dict podre ver todas las claves de mi diccionario esperando sus valores
```

```python
for line in data3: #Se comienza haciendo un recorrido de cada una de las líneas del data set
    values = line.strip().split(",") # de la primera línea se eliminan los espacios y se utilza la coma (,) como separador de cada valor de columna y con esto se crea un array
    for i in range(len(cols)): # se realiza un ciclo con el numero de columnas
        main_dict[cols[i]].append(values[i]) #se agrega el valor a su correspondiente clave del diccionario tomandola del array de values
    counter += 1 # se aumenta el contador que va contando las filas creadas
print('El data set tienen %d filas y %d columnas' % (counter ,n_cols))
```

```python
df3 = pd.DataFrame(main_dict) # Creamos el dataframe con el diccionario que se creo en la manipulación de los datos
df3.head()
```

## Lectura y escritura de ficheros

```python
infile = mainpath + "/" + "customer-churn-model/Customer Churn Model.txt"
outfile = mainpath + "/" + "customer-churn-model/Tab Customer Churn Model.txt"
```

```python
with open(infile,"r") as infile1: # abriendo el fichero infile en modo lectura y colocando como alias infile1
    with open(outfile, "w") as outfile1: # Y abriendo el archivo outfile en modo escriturau llaamando con el alias outfile1
        for line in infile1: # Realizo un cilco en cada linea del archivo infile1
            fields = line.strip().split(',') # leo cada linea y la separo por comas en un array
            outfile1.write('\t'.join(fields)) # escribo cada valor en una linea pero ahora los separa con la barra
            outfile1.write('\n') #Salto al linea al final de cada fila
```

# Ejercicio con descarga de datos con urllib
Este es un ejemplo usando la libreria urllib para leeer los datso de una url externa, procesando y convirtuiendolos en un dtaframe de *python* antes de guardarlos en un CVS local

```python
import csv
import urllib.request
```

```python
request_url = urllib.request.urlopen(medals_url)
response = request_url.read()
```

```python
def downloadFromURL(url, filename, sep = ",", delim = "\n", encoding="utf-8", 
                   mainpath = '/home/eligiuz/python-ml-course/datasets'):
    #primero importamos la librería y hacemos la conexión con la web de los datos
    request_url = urllib.request.urlopen(medals_url)
    response = request_url.read()
    
    #El objeto reponse contiene un string binario, así que lo convertimos a un string descodificándolo en UTF-8
    str_data = response.decode(encoding)

    #Dividimos el string en un array de filas, separándolo por intros
    lines = str_data.split(delim)

    #La primera línea contiene la cabecera, así que la extraemos
    col_names = lines[0].split(sep)
    n_cols = len(col_names)

    #Generamos un diccionario vacío donde irá la información procesada desde la URL externa
    counter = 0
    main_dict = {}
    for col in col_names:
        main_dict[col] = []

    #Procesamos fila a fila la información para ir rellenando el diccionario con los datos como hicimos antes
    for line in lines:
        #Nos saltamos la primera línea que es la que contiene la cabecera y ya tenemos procesada
        if(counter > 0):
            #Dividimos cada string por las comas como elemento separador
            values = line.strip().split(sep)
            #Añadimos cada valor a su respectiva columna del diccionario
            for i in range(len(col_names)):
                main_dict[col_names[i]].append(values[i])
        counter += 1

    print("El data set tiene %d filas y %d columnas"%(counter-1, n_cols))

    #Convertimos el diccionario procesado a Data Frame y comprobamos que los datos son correctos
    df = pd.DataFrame(main_dict)
    print(df.head())

    #Elegimos donde guardarlo (en la carpeta athletes es donde tiene más sentido por el contexto del análisis)
    fullpath = os.path.join(mainpath, filename)

    #Lo guardamos en CSV, en JSON o en Excel según queramos
    df.to_csv(fullpath+".csv")
    df.to_json(fullpath+".json")
    df.to_excel(fullpath+".xls")
    print("Los ficheros se han guardado correctamente en: "+fullpath)
    
    return df
```

```python
medals_df = downloadFromURL(medals_url, "athletes/downloaded_medals2")
medals_df.head()
```
